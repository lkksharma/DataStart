The following inputs and plots data from an SQL database

```python
import sqlite3 as sq3
import pandas.io.sql as pds
import pandas as pd
```

```python
path = 'data/classic_rock.db'
con = sq3.Connection(path)
```

```python
query = '''SELECT * FROM rock_songs'''
dataf = pds.read_sql(query, con)

dataf.iloc[:-1,:-1]
```

```python
query = '''SELECT Artist, Release_Year FROM rock_songs GROUP BY Artist, Release_Year'''

ord_data = pds.read_sql(query, con, parse_dates=['Release_Year'], chunksize = 10)

for index, row in enumerate(ord_data):
    if index<10:
        print(f'Chink number: {index}')
        display(row)
```

Forming plot to show the hike in western music by the Artists in the database

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
ax = plt.axes()
ax.hist(dataf.Release_Year, bins=100)

ax.set(xlabel = 'Year', ylabel = 'No. of Songs', title = 'Songs by Baby Boomer Artists')
```
